```python
import pandas as pd

metadata = pd.read_csv('panel-metadata.csv')
```

```python
metadata.head()
```

```python
_metadata = metadata.drop(columns=['Color', 'Unnamed: 3', 'Color2'], inplace=False)
_metadata = _metadata.dropna('index', how="all")
_metadata
```

```python
tagcol1= _metadata.iloc[:, 0]
tagcol2 = _metadata.iloc[:, 2]

tags = tagcol1.append(tagcol1)

hexcol1= _metadata.iloc[:, 1]
hexcol2 = _metadata.iloc[:, 3]

hexcode = hexcol1.append(hexcol2)

tags.index = range(len(tags))
hexcode.index = range(len(hexcode))

# hexcode.index = range(len(tags)) + 

# hexcode.index = range(len(tags), len(tags)+len(hexcode))

# print(range(len(tags)), range(len(tags), len(tags)+len(hexcode)))

# df = None
# # hexcode = hexcol1 

# # tagcol1.index

df = pd.DataFrame(
    {'tag': tags,'hexcode': hexcode}
    # index = range(len(tags))
)

df
```

```python
len(df.index)
_df = df.copy()

_df.columns = ['tag', 'rgb']
# _df
# rgb = []
# # colormaps = {}
for i in df.index:
    hex = df.iloc[i, 1]
    # print(hex)
    
    hex = str(hex)
    if len(hex) == 5:
        hex  = "0" + hex
    # print(len(str(hex)))
    assert len(hex ) == 6 or hex in ["NaN", "nan"], hex

    if hex not in ["NaN", "nan"]:
        rgb = tuple(int(str(hex)[j:j+2], 16) for j in (0, 2, 4))
    else:
        rgb = (0,0,0)

    _df.iloc[i, 1] = rgb

_df
```
